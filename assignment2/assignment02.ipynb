```python
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
from matplotlib import pyplot as plt
df=pd.read_csv('cinema_ticketing1.csv',encoding ='gbk') 
df
```

# 1. How many seats each cinema provides?

```python
from pylab import mpl 
mpl.rcParams[‘font.sans-serif] = [‘SimHei’]
```

```python
df1=df.drop_duplicates(['Cinema','Total Seats'],keep='first')
x=[0,2,3,4]
df1.drop(df1.columns[x], axis=1,inplace=True)
df1.groupby('Cinema').sum()
df1.plot('Cinema','Total Seats',color='#4169E1',kind='bar')
plt.rcParams['font.sans-serif']=['SimHei']
```

# 2. When did people tend to go for films more frequently?

```python
a=df['Showtime'].str.split(' ',expand=True)[4]
b=df['Sold Seats']
for i in range(0,68):
    a[i]=pd.to_datetime(a[i]).strftime('%H:%M')
data={'Time':a,'Sold Seats':b}
df2=DataFrame(data).groupby('Time').sum()
df2
```

```python
df2.plot()
```

# 3. The popularity of the latest movies on each cinemas?

```python
pd_df4 = df.pivot_table(index=['Cinema'], columns=['Movie Name'], values='Sold Seats', aggfunc='sum')
#pd_df4 
pd_df4 = pd_df4.loc[['MOViE MOViE Cityplaza','MY CINEMA YOHO MALL','PALACE apm','PALACE ifc','The ONE','數碼港','旺角','荃灣','荷里活','電影中心']].T
pd_df4.style.bar(color='#4169E1')
```
