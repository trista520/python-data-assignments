```python
#url='https://www.cinema.com.hk/tc/ticketing/seatplan/20012950'
import requests
import bs4
import csv
from bs4 import BeautifulSoup
```

```python
def scrape_one_page(url):
    info=[]
    page=requests.get(url)
    page.encoding = 'utf-8'
    mypage = BeautifulSoup(page.text,'lxml')
    
    name=mypage.find('div',attrs={'class':'movieName'})
    cinema=mypage.find('div',attrs={'class':'showInfoValue'})
    showtime=mypage.find_all('div',attrs={'class':'showInfoValue'})[2]
    
    q=mypage.find('div',attrs={'id':'main'})
    a_s = q.find_all('span')
    data = []
    for i in a_s:
        data.append(i['class'])
    del data[-1]
    delete=['block', 'preview', 'correridor', 'available', '']
    while delete in data:
        data.remove(delete)
        global condition
        condition=[]
        for word in data:
            condition.append(word[-2])
    available=condition.count('available')
    sold=condition.count('sold')
    if sold>30:
        info.append(name.text)
        info.append(cinema.text)
        info.append(showtime.text)#前三列
        info.append(available)    
        info.append(sold)
        info.append(len(condition))  
    return info

def get_all_pages(cinema_pages):
    all_info=[]
    for one_page in cinema_pages:
        info=scrape_one_page(one_page)
        all_info.append(info)
        while [] in all_info:
            all_info.remove([])
        if len(all_info)>=150:
            break
    return all_info
```

```python
list=[21111706,
      19128053,
      10146606,
      2236265,
      16232181]
cinema=[]
pages=[]
for i in list:
    one_cinema=range(i-30,i+100)     
    cinema.append(one_cinema)
cinema_number = [x for j in cinema for x in j]
base_url = 'https://www.cinema.com.hk/tc/ticketing/seatplan/'
for i in cinema_number:
    one_page=base_url+str(i)
    pages.append(one_page)
all_info=get_all_pages(pages)
```

```python
with open('cinema_ticketing1.csv','w',newline='') as g:
    writer = csv.writer(g)
    header = ['Movie Name','Cinema','Showtime','Available Seats','Sold Seats','Total Seats']
    writer.writerow(header)
    writer.writerows(all_info)
```
